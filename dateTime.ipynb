```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

dates = pd.Series(pd.date_range("2012-02-04", "2012-06-14"))
```

```python
df = pd.DataFrame(dates, index=dates)
df.head()
```

```python
df.index.month == 3
```

```python
march = df[df.index.month == 3]
march
```

```python
df1 = pd.DataFrame(dates, columns=['Dates'])
df1.head()
```

```python
df1['Year'] = df1['Dates'].map(lambda x: x.year)
df1['Month'] = df1['Dates'].map(lambda x: x.month)
df1.head()
```

```python
marchFilter = df1['Month'] == 3
march = df1[marchFilter]
march = march['Dates']
march
```
