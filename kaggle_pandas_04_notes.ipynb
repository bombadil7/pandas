```python
import pandas as pd
import numpy as np

reviews = pd.read_csv("data/winemag-data-130k-v2.csv", index_col=0)
pd.set_option("display.max_rows", 5)

reviews.head()
```

### Grouping
Use __groupby()__ the same as __reviews.points.value_counts()__

```python
pd.set_option("display.max_rows", 10)
reviews.groupby('points').points.count()
```

To get the cheapest wine in each wine category we can do the following:

```python
reviews.groupby('points').price.min()
```

This makes each category act as a DataFrame.  We can use __apply__ method on these categories.
To get the name of the first wine reviewed from each winery:

```python
reviews.groupby('winery').apply(lambda df: df.title.iloc[0])
```

We can filter by more than one category:

```python
pd.set_option("display.max_rows", 5)
reviews.groupby(['country', 'province']).apply(lambda df: df.loc[df.points.idxmax()])
```

Another __groupby()__ method to know is __agg__, which lets you run several methods on the DataFrame simultaneously. 

```python
reviews.groupby(['country']).price.agg([len, min, max])
```

### Multi-Indexes
__groupby__ can create multi-index:

```python
countries_reviewed = reviews.groupby(['country', 'province']).description.agg([len])
countries_reviewed
```

Multi-index has multiple levels.  They require two levels of labels to retrieve a value.  
The most common method is for converting back to regular index - __reset_index()__

```python
countries_reviewed.reset_index()
```

### Sorting
__groupby__ sorts items by the index instead of data.  To sort differently use __sort_values__:

```python
countries_reviewed = countries_reviewed.reset_index()
countries_reviewed.sort_values(by='len')
```

Change from assending sort to descending:

```python
countries_reviewed.sort_values(by='len', ascending=False)
```

__sort_index__ is the companion method to sort in default order:

```python
countries_reviewed.sort_index()
```

We can sort by more than one column at a time:

```python
countries_reviewed.sort_values(by=['country', 'len'])
```
